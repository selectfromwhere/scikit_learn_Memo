# 決定木分析とは
## 概要
- 意思決定ツリー、デシジョンツリーとも呼ばれる教師ありの機械学習手法
- 樹形図状のモデル（ノード(節)と、そこから分岐するリーフ(葉)で示される）を使い、購買履歴などの結果が記録されたデータセットを**分類**する
- 分類結果に影響を与えた要因を分析し、そのモデルを利用して将来の予測を行う

### 識別(分類)と回帰とは
Web検索だと書き手によって微妙に解釈が異なり、ちょっと解釈が難しかった。  
ここは様々な人の目に晒されているWikipediaから、理系学部出身だが文系脳の私が自分なりに噛み砕いて解釈して見る。  

識別(分類)と回帰のどちらも統計学において、データを複数のクラス（グループ）に*分類するための線引きをすること*を指す。  

#### 私的な解釈のポイント
- 分類(識別)と回帰と表現されるとめちゃくちゃ混乱する（クラスに分類するための方法に、分類と回帰があるとか何言ってだ）  
- 回帰と回帰分析を分けて考えないとまあまあ混乱する（回帰で分類するのと、回帰で得られたモデルで予測を行うのは別の話）

#### 識別(分類)
**離散値**の目的変数をクラス分けすることを**識別(分類)**という。
どちらかというと、識別と書いた方がわかりやすい。  
決定木分析では、訓練データから分類器を生成し、その分類器を利用してデータを分類していく。

#### 回帰
**連続値**の目的変数をクラス分けすることを**回帰**という。  
目的変数と説明変数の間にモデルを当てはめて、目的変数をクラスに分類する。  
回帰分析では、目的変数と説明変数の間にモデルを当てはめたモデルを利用して数値を予測していく。

例）



### 連続値と離散値
連続値とも離散値とも取れる場合があって、ちょっとだけ混乱するけどまあ大丈夫。


#### 連続値
測ることができるデータ、連続していて細かく測ることができる。  
数直線上に書くと、どこにでも存在しうるデータ

例）身長、体重、時間、気温など


#### 離散値
数えることができるデータ、それ以上細かくすることができない。  
数直線上に書くと、目盛の上に存在するデータ

例）人数、回数など


#### どちらとも取れるケース
・連続値の場合  
身長や体重は、kg単位で表されることから離散値とも考えられる（数えられるデータ）  
便宜的に近似値に直しているとして、基本的には離散値として扱うことが多いようだ。

・離散値の場合  
テストの点数をその人の能力を階級値で示すものと考え、75点は74.5〜75.4点を取れる能力とし、連続値として扱う場合があるようだ。

### 決定木分析の特徴

> * カテゴリカルデータと数値データの両方を扱うことが可能
> * 検定を行って、作成したモデルの正しさを評価することが可能

## 試行過程と結果
### データセットを確認する

[Datasetはここから入手した](https://vincentarelbundock.github.io/Rdatasets/datasets.html)

```python
# 編集前のcsvファイルを読み込んで、DataFrame化
import pandas as pd
import_df = pd.read_csv('HairEyeColor.csv')
import_df.head()
```

決定木分析をするにあたり、このデータセットでは２工程の下処理が必要
- 各レコードをFreqの数だけ生成する
- String型の各項目をint型に置換する

### １.CSVデータを加工する

```python
in_file  = open("HairEyeColor.csv","r")
out_file = open("HairEyeColor_Edited.csv","w")

# アウトプットファイルにヘッダーを書き込み
out_file.write("Hair,Eye,Sex\n")

# インプットファイルのヘッダーを読み飛ばす
in_file.readline()
# インプットファイルの全レコードを読み込み
lines = in_file.readlines()

# for文で1行ずつ処理
for line in lines:
    # 改行コードはブランクに置換
    line = line.replace("\n","")
    # カンマ区切りでリストに変換する
    line = line.split(",")
    # 変換処理した値を、更にカンマ区切りへ変換
    row = "{},{},{}\n".format(line[1],line[2],line[3])
    # 書き出し用のファイルに"Freq"の数だけ出力
    freq = int(line[4])
    for i in range(0,freq):
        out_file.write(row)

in_file.close()
out_file.close()
```

```python
# 編集後のcsvファイルを読み込んで、DataFrame化
import_df = pd.read_csv('HairEyeColor_Edited.csv')
import_df.head()
```

```python
# 加工前との整合性を確認する
check = import_df.groupby(['Hair','Eye','Sex']).size()
check
```

```python
# データの特徴を見てみる
import_df.describe()
```

```python
# String型の各項目をint型に置換するための辞書を作って変換
dict = {'Hair':{'Black' :1,'Blond':2,'Brown':3,'Red'  :4},
        'Eye' :{'Blue'  :1,'Brown':2,'Green':3,'Hazel':4},
        'Sex' :{'Female':1,'Male' :2}}

import_df['Hair'] = import_df["Hair"].map(dict['Hair'])
import_df['Eye']  = import_df["Eye"].map(dict['Eye'])
import_df['Sex']  = import_df["Sex"].map(dict['Sex'])
import_df.head()
```

### ２.決定木の分類器を作成して可視化する

```python
from sklearn import tree
# 説明変数は'Hair','Eye'
variables = ['Hair','Eye']

# 決定木の分類器を作成
classifier = tree.DecisionTreeClassifier()

# 目的変数は'Sex'
# サンプルデータを食わせて学習
classifier = classifier.fit(import_df[variables],import_df['Sex'])

# 学習結果を.dotファイルで出力する方法
#  from sklearn.tree import export_graphviz
#  with open('graph.dot','w') as f:
#     f = tree.export_graphviz(classifier, out_file=f)
```

```python
# 作成した決定木を可視化する
import pydotplus
from sklearn.externals.six import StringIO
dot_data = StringIO()
tree.export_graphviz(classifier, out_file=dot_data,
                     filled=True,rounded=True)

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())

from IPython.display import Image
Image(graph.create_png())
```

### ３.交差検証
今回は決定木の深さだけを変えて交差検証してみる

```python
# 交差検証１（max_depth指定なし）
import numpy as np
from sklearn import cross_validation as cv

data = import_df.reindex(np.random.permutation(import_df.index))
variables = ['Hair','Eye']

classifier = tree.DecisionTreeClassifier()
scores = cv.cross_val_score(classifier, data[variables], data['Sex'], cv=5)

print(scores.mean(), scores)
```

```python
# 交差検証２（max_depth=3）
data = import_df.reindex(np.random.permutation(import_df.index))
variables = ['Hair','Eye']

classifier = tree.DecisionTreeClassifier(max_depth=3)
scores = cv.cross_val_score(classifier, data[variables], data['Sex'], cv=5)

print(scores.mean(), scores)
```

```python
# 交差検証３（max_depth=4）
data = import_df.reindex(np.random.permutation(import_df.index))
variables = ['Hair','Eye']

classifier = tree.DecisionTreeClassifier(max_depth=4)
scores = cv.cross_val_score(classifier, data[variables], data['Sex'], cv=5)

print(scores.mean(), scores)
```
